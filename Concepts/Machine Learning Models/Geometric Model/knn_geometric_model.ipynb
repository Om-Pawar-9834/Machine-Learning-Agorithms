```python
from sklearn.neighbors import KNeighborsClassifier

X = [
    [150, 8],
    [160, 7],
    [170, 8],
    [120, 5],
    [130, 4],
    [140, 5]
]

y = [
    "Apple",
    "Apple",
    "Apple",
    "Orange",
    "Orange",
    "Orange"
]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

new_fruit = [[155, 8]]
prediction = model.predict(new_fruit)

print("Predicted Fruit:", prediction[0])
```

# Code Explaination

# Imports the KNeighborsClassifier class from the sklearn.neighbors module so that we can create and use the K-Nearest Neighbors (KNN) classification model in our program.

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Training data
# features:[weight(grams) , sweetness(1-10)]

```python
x = [
    [150, 8],   # Apple
    [160, 7],   # Apple
    [170, 8],   # Apple
    [120, 5],   # Orange
    [130, 4],   # Orange
    [140, 5]    # Orange
]
```

# Label (actual fruit names)

```python
y = [
    "Apple",
    "Apple",
    "Apple",
    "Orange",
    "Orange",
    "Orange"
]
```

# Create KNN model 
# n_neighbors=3 means:
# The model will look at the 3 nearest fruits
# before making a prediction.

```python
model = KNeighborsClassifier(n_neighbors=3)
```

# Train the model
# The model stores the training data
# so it can compare future data with it.

```python
model.fit(x,y)      #the fit() method returns the trained model object. Since it is the last line in the cell, Jupyter automatically displays its representation.
```

# New Fruit
# Weight = 155 grams
# Sweetness = 8

```python
new_fruit=[[155,8]]
```

# Predict

```python
prediction = model.predict(new_fruit)
```

# Display Result

```python
print("Prediction : ",prediction[0])  
#prediction[0] is used to get the first predicted result because model.predict() always returns an array, even for a single prediction
```
